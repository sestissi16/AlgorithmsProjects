```python
def DictOfNodes(file):
    dict = {}
    nodes = []
    firstTime = True
    firstedge= ''
    secondedge = '' 
    filename = open(file, 'r')
    filen = filename.read()
    print(filen)
    for node in filen.split():
        nodes.append(node)
    for node in nodes:
        print(node)
        for edge in node.split('-'):
            print(edge)
            if firstTime:
                if edge in firstedge:
                    firstTime = False
                else:
                    dict[edge] = []
                    firstedge = edge
                    firstTime = False
                print(dict)
            else:
                secondedge = edge
                dict[firstedge].append(secondedge)
                firstTime=True
                
    print(dict)
                
            
            
            
            

DictOfNodes("graphSmall1.in")
```
